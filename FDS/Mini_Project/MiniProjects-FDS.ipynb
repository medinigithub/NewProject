```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diamonds=pd.read_csv('D:\diamonds.csv',names=['id','carat','cut','color','clarity','depth','table','price','x','y','z'])
print(diamonds.head())
```

```python
fig,ax=plt.subplots()
ax.scatter(diamonds['depth'],diamonds['price'])
ax.set_title('Diamonds Dataset')
ax.set_xlabel('depth')
ax.set_ylabel('price')
```

```python
fig,ax=plt.subplots()
ax.hist(diamonds['price'])
ax.set_title('Diamonds_dataset')
ax.set_xlabel('price')
ax.set_ylabel('Freuency')
```

```python
wine_reviews=pd.read_csv('D:\winemag-data.csv',index_col=0)
wine_reviews.head()
```

```python
fig,ax=plt.subplots()
data=wine_reviews['points'].value_counts()
points=data.index
frequency=data.values
ax.bar(points,frequency)
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
wine_reviews['points'].value_counts().sort_index().plot.bar()
```

```python
wine_reviews.groupby("country").price.mean().sort_values(ascending=False)[:5].plot.bar()
```
